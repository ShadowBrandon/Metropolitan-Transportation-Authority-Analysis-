```python
adjtime = [] #Make an empty set for me and copy the time to an empty set

#OBJ: By subtracting time from the original one, we can consider the appropriate set
#Simple 
#0        2018-06-22 23:55:00    from 2018-06-23 00:00:00
#1        2018-06-23 03:55:00    from 2018-06-23 04:00:00
#2        2018-06-23 07:55:00
#3        2018-06-23 11:55:00
#4        2018-06-23 15:55:00
#5        2018-06-23 19:55:00      does not consider the last FOUR HOURS :(
#6        2018-06-23 23:55:00  from   2018-06-24 00:00:00 -- next day :(
################ WE NOTE THAT WE COUNT FOR EVERY FOUR HOURS!!! ##################
```

```python
#correcttime = adjtime - pd.Timedelta(minutes=5) 
#correcttime.head(10) #Time moved back 5 minutes (NOT UNIQUE)
#pd.Timedelta() is a funct time converter
#Take my copied time and subtract 5 minutes and store this into correct time 
```

```python
turnstile_att1["DATE/TIME"] = adjtime - pd.Timedelta(minutes=5) 
turnstile_att1['DATE/TIME'] = pd.to_datetime(turnstile_att1['DATE/TIME']).dt.date #GET RID OF TIME I WANT DATE!!!
turnstile_att1.head(5)
turnstileset = turnstile_att1.iloc[::-1] 
turnstileset.head(100)

#### TURNSTILE_ATT1 -- REPLACED INCORRECT TIME WITH CORRECT TIME!!!
```

```python
# EntriesDif=turnstileset["ENTRIES"].diff() #Diff does the following C1[i] = C[i] - C[i-1] column wise Don't respect grouping :(
EntriesDif=turnstileset.groupby(by=["DATE/TIME", "C/A", "UNIT", "SCP", "STATION"]).ENTRIES.diff()

# 1 --> 6667173 NOW 1 = 6667173 [[1]] - 6667150 [[2]]
# 2 --> 6667189 NOW 1 = 6667189 [[2]] - 6667173 [[1]]
#NOTES: 0 is only day therefore nan and 1 is first of next dat therefore nan 
EntriesDif.head(100) #Just call you call it don't make permanent change (i+1) != (i=i+1)
```

```python
turnstileset1= turnstileset[["C/A", "UNIT", "SCP", "STATION", "DATE/TIME"]]
turnstileset1["ENTRIES"] = EntriesDif #Please remember that this considers time AND LOCATION (a || b1 b2 b3 || c1 c2 c3)
turnstileset1.head(3) #Turnstileset1 has all the most updated things we need with JUST AND ONLY the date repeating
```
